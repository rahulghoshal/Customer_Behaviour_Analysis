```python
#Checking the null values in the dataset
data.isnull().sum()
```

```python
#Rows where Review Rating is null
null_reviews = data[data['Review Rating'].isna()]
null_reviews
```

```python
#Null rating in each category
null_reviews["Category"].value_counts()
```

```python
#Replacing the Null values of each category with median rating of each category
data["Review Rating"] = data.groupby("Category")["Review Rating"].transform(lambda x: x.fillna(x.median()))
data.isnull().sum()
```

```python
#Formatting the column names
print(f"Before formatting the column names : \n{data.columns}")
data.columns = data.columns.str.lower()
data.columns = data.columns.str.replace(" ","_")
print(f"After formatting the column names : \n{data.columns}")
```

```python
data = data.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
data.columns
```

```python
#Creating a column age_group
labels = ["Young Adult","Adult","Middle Aged","Senior"]
data["age_group"] = pd.qcut(data["age"],q=4,labels=labels)
```

```python
data[["age","age_group"]]
```

```python
#Creating a column purchase_freqency_days
data["frequency_of_purchases"].unique()
```

```python
freq = {
        "Fortnightly" : 14,
        "Weekly" : 7,
        "Annually" : 365,
        "Quarterly" : 90,
        "Bi-Weekly" : 14,
        "Monthly" : 30,
        "Every 3 Months" : 90
        }
data["purchase_freqency_days"] = data["frequency_of_purchases"].map(freq)
data[["frequency_of_purchases","purchase_freqency_days"]]
```

```python
#Checking if the columns "promo_code_used" and "discount_applied" is same or not
(data["promo_code_used"] == data["discount_applied"]).all()
```

```python
data = data.drop("promo_code_used",axis=1)
```
